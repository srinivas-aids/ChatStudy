```python
dfc=pd.read_csv('student_score.csv')
dfc
```

```python
sdmerge=df.merge(dfc,on='sid')
sdmerge
```

```python
x=sdmerge['mentorscore']*10
y=sdmerge['progscore']*5
z=sdmerge['course staff score']*20
sdmerge['sd']=(x+y+z)
```

```python
sdmerge['sd']
```

```python
q3=sdmerge['sd'].quantile(0.75)
q3
```

```python
q1=sdmerge['sd'].quantile(0.25)
q1
```

```python
iqr=q3-q1
iqr
```

```python

ub=q3+(1.5*iqr)
ub
```

```python
lb=q1-(1.5*iqr)
lb
```

```python
ch=['Name','Score','Status']
```

```python
df=pd.DataFrame(columns=ch)
lb=200
df
```

```python
sdmerge
```

```python
def status(score,lb,ub):
  if score>=lb and score<=ub:
    return 'Selected'
  else:
    return 'not selected'
df['Status']=sdmerge['sd'].apply(lambda x:status(x,lb,ub))
df['Name']=sdmerge['sname']
df['Score']=sdmerge['sd']
df
```

```python
df.to_csv('selected.csv')
```

```python
import pandas as pd
import numpy as np
```

```python
afm=pd.read_csv('ahamovies.csv')
```

```python
afm.head()
```

```python
afc=pd.read_csv('ahacredits.csv')
```

```python
afc.head()
```

```python
afm=afm.rename(index=str,columns={"movieid" : "id"})
```

```python
afm.head()
```

```python
print("Movies:",afm.shape)
print("Survey:",afc.shape)
```

```python
afmg=afm.merge(afc,on='id')
afmg.head()
```

```python
afmg.dropna(axis=1,inplace=True)
afmg
```

```python
afmg.info()
```

```python
afmg.head()
```

```python
p=afmg['votecount']
q=afmg['voteavg']
r=afmg['voteavg'].mean()
s=afmg['votecount'].quantile(0.70)
```

```python
afmg['average']=((p*q)+(r*s))/(p+s)
```

```python
movierank=afmg.sort_values('average',ascending=False)
movierank[['movietitle','votecount','voteavg','average','cast']].head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
avg=movierank.sort_values('average',ascending=False)
plt.figure(figsize=(12,6))
axis1=sns.barplot(x=avg['average'].head(5),y=avg['movietitle'].head(5),data=avg,palette='bright')
plt.xlim(4,10)
plt.title('Best Movies by average votes',weight='bold')
plt.xlabel('Average Scores',weight='bold')
plt.ylabel('Movies Title',weight='bold')
plt.savefig('bestmovies.png')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
avg=movierank.sort_values('average',ascending=False)
plt.figure(figsize=(10,4))
axis1=sns.barplot(x=avg['average'].head(5),y=avg['movietitle'].head(5),data=avg,hue=avg['cast'].head(5))
plt.xlim(4,10)
plt.title('Best Movies by Average votes',weight='bold')
plt.xlabel('average Score',weight='bold')
plt.ylabel('movie title',weight='bold')
plt.savefig('best_movies.png')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
avg=movierank.sort_values('average',ascending=True)
plt.figure(figsize=(8,4))
axis1=sns.barplot(x=avg['average'].head(3),y=avg['movietitle'].head(3),data=avg,hue=avg['cast'].head(3))
plt.xlim(4,10)
plt.title('underrated by Average votes',weight='bold')
plt.xlabel('average Score',weight='bold')
plt.ylabel('movie title',weight='bold')
plt.savefig('underratedmovies.png')
```
